```python
import pandas as pd

# Cargar el dataset original
df = pd.read_csv("data/dataset_v1.csv")

# Eliminar filas con valores nulos
df_clean = df.dropna()

# Eliminar filas duplicadas
df_clean = df_clean.drop_duplicates()

# Guardar el nuevo dataset
df_clean.to_csv("data/dataset_v2.csv", index=False)

print("‚úÖ dataset_v2.csv creado exitosamente con datos limpios.")
```

```python
import pandas as pd

# Cargar datasets
df_v1 = pd.read_csv("data/dataset_v1.csv")
df_v2 = pd.read_csv("data/dataset_v2.csv")

# 1. Dimensiones
print("üìè Dimensiones:")
print(f"dataset_v1: {df_v1.shape}")
print(f"dataset_v2: {df_v2.shape}\n")

# 2. Nulos
print("üßº Valores nulos:")
print(f"dataset_v1: {df_v1.isnull().sum().sum()} nulos totales")
print(f"dataset_v2: {df_v2.isnull().sum().sum()} nulos totales\n")

# 3. Duplicados
print("üîÅ Filas duplicadas:")
print(f"dataset_v1: {df_v1.duplicated().sum()} duplicadas")
print(f"dataset_v2: {df_v2.duplicated().sum()} duplicadas\n")

# 4. Estad√≠sticas descriptivas
print("üìä Estad√≠sticas descriptivas (v1):")
print(df_v1.describe())

print("\nüìä Estad√≠sticas descriptivas (v2):")
print(df_v2.describe())

# 5. Diferencias en contenido
print("\nüîç Diferencias en contenido:")
diff_rows = pd.concat([df_v1, df_v2]).drop_duplicates(keep=False)
print(f"Filas diferentes entre v1 y v2: {diff_rows.shape[0]}")
```
