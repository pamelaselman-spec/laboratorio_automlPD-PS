```python
import pandas as pd
import numpy as np

# Cargar el dataset original
df = pd.read_csv("data/dataset_v3.csv")

# Número de registros nuevos que quieres generar
n_nuevos = 1000

# Crear nuevos registros sintéticos
df_nuevos = pd.DataFrame()

for col in df.columns:
    if df[col].dtype == "object":
        # Para columnas categóricas: muestreo aleatorio de valores existentes
        df_nuevos[col] = np.random.choice(df[col].dropna().unique(), size=n_nuevos)
    else:
        # Para columnas numéricas: muestreo normal alrededor de la media
        media = df[col].mean()
        std = df[col].std()
        df_nuevos[col] = np.random.normal(loc=media, scale=std, size=n_nuevos).round(2)

# Concatenar con el dataset original
df_ampliado = pd.concat([df, df_nuevos], ignore_index=True)

# Guardar el nuevo dataset ampliado
df_ampliado.to_csv("data/dataset_v3.csv", index=False)
print("✅ Dataset ampliado guardado como data/dataset_v3.csv")
```
